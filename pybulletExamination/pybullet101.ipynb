```python
import pybullet as p
import time
import pybullet_data
#https://medium.com/@chand.shelvin/pybullet-getting-started-a068a0e3d492
# 한국어 nice intro설명 : https://blog.naver.com/PostView.nhn?blogId=einsbon&logNo=221549883110&parentCategoryNo=&categoryNo=1&viewDate=&isShowPopularPosts=true&from=search
connect = False
```

```python
if not connect:
    physicsClient = p.connect(p.GUI)
    connect = True
```

```python
p.setAdditionalSearchPath(pybullet_data.getDataPath()) #pybullet_data.getDataPath()는 pybullet이 기본적으로 제공하는 데이터 파일(땅, 몇 가지 로봇 등)이 있는 경로입니다.


p.setGravity(0,0,-10)
#이 숫자를 조절하는 것으로, 물리 시뮬레이션에서 한번 업데이트할 때 몇 초 업데이트할지를 설정할 수 있습니다. 
# 타임스텝을 1/200 으로 설정하면 200번 업데이트해야 그 시뮬레이션 세계에서 1초가 흐른 것입니다(초당 200프레임). 
# 숫자가 작을수록 정밀하게 시뮬레이션되지만, 어떤 시간만큼 시뮬레이션 했을 때 더 많은 스텝이 필요하겠죠.

planeId = p.loadURDF("plane.urdf")
cubeStartPos = [2,0,1]
cubeStartOrientation = p.getQuaternionFromEuler([0,0,3.14]) 
# loadURDF 함수는 위치와 회전 상태도 설정할 수 있습니다. 회전 상태는 쿼터니언 방식으로 인자를 받기 때문에 10번 줄에서 오일러 각에서 쿼터니안 형식으로 바꾼 것입니다.
boxId = p.loadURDF("r2d2.urdf",cubeStartPos, cubeStartOrientation)
```

```python
for i in range (1000):
   p.stepSimulation()
   time.sleep(1./240.)
   print("position/orientation : ", p.getBasePositionAndOrientation(boxId))
# stepSimulation으로 시뮬레이션을 한번 진행시킬 수 있습니다. 여기서는 계속 진행시키기 위해 루프 안에서 반복시켰습니다. 
# sleep을 안 쓰면 너무 빠르게 진행되어서 sleep으로 시뮬레이션을 더 보기 좋게 만들었습니다. 
# 그리고 stepSimulation도 수행하는 데 시간이 걸리기 때문에 time.sleep으로 설정한 타임스텝 만큼 쉬어도 시뮬레이션의 보여주기는 현실보다 더 느립니다.
# [출처] PyBullet 물리 시뮬레이션 기초 사용법 강좌|작성자 선빈
cubePos, cubeOrn = p.getBasePositionAndOrientation(boxId)
print(cubePos,cubeOrn)
```

```python
import pybullet as p
import time
import pybullet_data
physicsClient = p.connect(p.GUI)
p.setAdditionalSearchPath(pybullet_data.getDataPath())
p.setGravity(0, 0, -9.8)
p.setTimeStep(1/200)
planeID = p.loadURDF('samurai.urdf')
hndStartPos = [2, -1, 1]
hndStartOrientation = p.getQuaternionFromEuler([0, 0, 3.14])
humanoid3 = p.loadURDF("humanoid/humanoid.urdf",
                       hndStartPos, hndStartOrientation,
                       globalScaling=0.25,
                       useFixedBase=False)
run_simul = False
if run_simul:
    for i in range(10000):
        p.stepSimulation()
        time.sleep(1 / 200)
    p.disconnect(physicsClient)
```

```python
p.getJointInfo(humanoid3, 0)
# (0, #jointindex
#  b'root', #jointName
#  4, #jointType
#  -1, #qIndex: the first position index in the positional state variables for this body
#  -1, #uIndex: the first velocity index in the velocity state variables for this body
#  0, #flags
#  0.0, #jointDamping
#  0.0, #jointFriction
#  0.0, #jointLowerLimit
#  -1.0, #jointUpperLimit
#  0.0, #jointMaxForce
#  0.0, #jointMaxVelocity
#  b'root', #linkName
#  (0.0, 0.0, 0.0), #jointAxis : joint axis in local frame (ignored for JOINT_FIXED)
#  (0.0, 0.0, 0.0), #parentFramePos
#  (0.0, 0.0, 0.0, 1.0), #parentFrameOrn
#  -1) #parentIndex : parent link index, -1 for base
p.getJointInfo(humanoid3, 1)
```
